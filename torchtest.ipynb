```python
%config Completer.use_jedi=False
import torch
from torch import nn
import numpy as np
from torch import optim
from torch.autograd import Variable
import torch.nn.functional as F
```

```python
torch.cuda.is_available()
```

```python
class NeuralNet(nn.Module):
    
    def __init__(self):
        super(NeuralNet,self).__init__()
        self.fc1=nn.Linear(1,10)
        self.fc2=nn.Linear(10,1)

    def forward(self,x):
        x=F.relu(self.fc1(x))
        out=self.fc2(x)
        return out
```

```python
# train=torch.rand([25,10])
train=np.array(range(1,10001))*10
mean=np.mean(train)
std=np.std(train)
train=train - mean
train=train/std
train=torch.from_numpy(train)
train=train.view(-1,1)
train=train.type('torch.FloatTensor')
```

```python
target=train*23


train, target = Variable(train), Variable(target)
```

```python
x = torch.unsqueeze(torch.linspace(-1, 1, 100), dim=1)  # x data (tensor), shape=(100, 1)
y = x.pow(2) + 0.2*torch.rand(x.size())                 # noisy y data (tensor), shape=(100, 1)

# torch can only train on Variable, so convert them to Variable
train, target = Variable(x), Variable(y)
```

```python
net=NeuralNet()
```

```python
# y.backward(torch.rand(y.shape))
# target=torch.rand(y.shape)
```

```python
criterion=nn.MSELoss()
```

```python
optimizer=optim.SGD(net.parameters(),lr=0.2)
```

```python
print('net.fc1.weight',net.fc1.weight)
print('net.fc1.weight.grad',net.fc1.weight.grad)

print('net.fc1.bias',net.fc1.bias)
print('net.fc1.bias.grad',net.fc1.bias.grad)
```

```python
for _ in range(1000):
    optimizer.zero_grad()
#     print('\n\n Zero the gradiants')
#     print('net.fc1.weight',net.fc1.weight)
#     print('net.fc1.weight.grad',net.fc1.weight.grad)

#     print('net.fc1.bias',net.fc1.bias)
#     print('net.fc1.bias.grad',net.fc1.bias.grad)
    y=net.forward(train)
    loss=criterion(target,y)
    print('\nLoss',loss)
    loss.backward()
    optimizer.step()
#     print('\n\n Update the weights')
#     print('net.fc1.weight',net.fc1.weight)
#     print('net.fc1.weight.grad',net.fc1.weight.grad)

#     print('net.fc1.bias',net.fc1.bias)
#     print('net.fc1.bias.grad',net.fc1.bias.grad)
```
